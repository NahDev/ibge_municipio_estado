# Atividade em Python para a empresa Mont Capital Asset

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import requests
import zipfile
import os
```

```python
estados_response = requests.get("https://servicodados.ibge.gov.br/api/v1/localidades/estados")
estados_data = estados_response.json()
```

```python

municipios_por_estado = {}
for estado in estados_data:
    estado_id = estado['id']
    municipios_response = requests.get(f"https://servicodados.ibge.gov.br/api/v1/localidades/estados/{estado_id}/municipios")
    municipios_data = municipios_response.json()
    municipios_por_estado[estado['sigla']] = municipios_data
```

```python

dados = []
for estado_sigla, municipios in municipios_por_estado.items():
    for municipio in municipios:
        dados.append({'Estado': estado_sigla, 'Município': municipio['nome']})
```

```python

df = pd.DataFrame(dados)
```

```python
df1 = df.sort_values(by=['Estado', 'Município']).reset_index(drop=True)
```

```python
df1

```

```python

municipios_por_estado = df['Estado'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
ax = municipios_por_estado.plot(kind='bar')

for i, v in enumerate(municipios_por_estado):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')

plt.title('Número de Municípios por UF')
plt.xlabel('UF')
plt.ylabel('Número de Municípios')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# print(f'Iniciando o processo de download do arquivio .shp das UFs {ufs}\n')

for uf in ufs:
    print(f'Baixando .shp de {uf}')
    url_zip = f"https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2022/UFs/{uf}/{uf}_Municipios_2022.zip"
    # print(url_zip)
    
    nome_arquivo_zip = f"{uf}_Municipios_2022.zip"
    diretorio_destino = "dados_geograficos"
    
    caminho_arquivo_zip = os.path.join(diretorio_destino, nome_arquivo_zip)
    os.makedirs(diretorio_destino, exist_ok=True)
    r = requests.get(url_zip, allow_redirects=True)
    open(caminho_arquivo_zip, 'wb').write(r.content)
    
    with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zip_ref:
        zip_ref.extractall(diretorio_destino)
    
    arquivos_extraidos = os.listdir(diretorio_destino)
    for arquivo in arquivos_extraidos:
        if not arquivo.endswith('.shp') and not arquivo.endswith('.shx'):
        # if not arquivo.endswith('.shx'):
            os.remove(os.path.join(diretorio_destino, arquivo))
    
    arquivos_finais = os.listdir(diretorio_destino)
print(f'Download do arquivo {arquivos_finais} realizado com sucesso!!!' )

    
```
